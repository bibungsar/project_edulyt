# Tasks

## Customer Segmentation

```python
def age_group(a,g):
    if g=='F':
        if a >= 18 and a <= 30:
            return 'Young Females'
        elif a>=31 and a<=45:
            return 'Mid-age Females'
        else:
            return 'Old Females'
    else:
        if a>=18 and a<=30:
            return 'Young Males'
        elif a>=31 and a<=45:
            return 'Mid-age Males'
        else:
            return 'Old Males'
```

```python
ci['Age_Group']=ci.apply(lambda x:age_group(x['Age'],x['Gender']),axis=1)
```

```python
ci.head()
```

## Segmentwise Spend in Dollars

```python
ci['Spend_Amt']=td.groupby('Credit_card')['Selling_price'].sum().reset_index()['Selling_price']
```

```python
ci['Swipes']=td.groupby('Credit_card').count().reset_index()['Selling_price']
```

```python
ci.head()
```

```python
ss=ci.groupby('Age_Group')['Spend_Amt'].sum().sort_values(ascending=False).reset_index()
```
